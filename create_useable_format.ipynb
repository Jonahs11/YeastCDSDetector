```python
from Bio import SeqIO
from Bio.SeqFeature import CompoundLocation
import numpy as np
import pandas as pd
```

```python
record = SeqIO.read("./yeast_chr16/chr16.gbf", "genbank")
chrom_name = record.id

# coding_mask = build_coding_mask(record)
```

```python
len(record.seq)
```

```python
# strand 1 pass

total_length = len(record.seq)
is_cds_mask = np.zeros((total_length, ))

desired_feature_type = "CDS"

for feature in record.features[:100]:
    if feature.type != desired_feature_type:
        continue
    location = feature.location
    if isinstance(location, CompoundLocation):
        groups = location.parts
    else:
        groups = [location]

    for group in groups:
        is_cds_mask[group.start:group.end] = 1
is_cds_mask = is_cds_mask.astype(np.bool)
```

```python
df = pd.DataFrame({"seq": np.array(record.seq), "is_cds": is_cds_mask})
```

```python
df.to_csv("./seq_to_coding_file.csv", sep=",", index=False)
```
